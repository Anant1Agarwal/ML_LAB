```python
import pandas as pd 
import numpy as np 
```

```python
data=pd.read_csv('patient_data.csv')
data.head()
data=data.drop('Patient ID',axis=1)
data
```

```python
def classify_age(age):
    if age <=30:
        return 'young'
    elif 30 < age <= 50:
        return 'mid'
    else:
        return 'old'
```

```python
data['Age'] = data['Age'].astype(int)
data['Age']=data['Age'].apply(classify_age)
data
```

```python
class Node: 
    def __init__(self,):
        self.attribute=None
        self.leaf_node=False
        self.branches={}
        self.ans=None

class ID3DecisionTree:
    def __init__(self):
        self.root=None

def calc_info_gain(data,feature, target): 
    entropy_parent=0 
    for cls_label in data[target].unique(): 
        p_cls=len(data[data[target]==cls_label])/len(data)
        if p_cls==0: 
            continue
        entropy_parent+=-p_cls*np.log2(p_cls)

    entropy_children=0 
    for x in data[feature].unique(): 
        entropy_subset=0
        for y in data[target].unique(): 
            p_cls_branch=len(data[(data[feature]==x) & (data[target]==y)])/len(data[data[feature]==x])
            if p_cls_branch==0: 
                continue  
            entropy_subset+=-p_cls_branch*np.log2(p_cls_branch)
        entropy_children+=entropy_subset*(len(data[data[feature]==x])/len(data))

    return entropy_parent-entropy_children

def recursive_split(data,target): 
    if len(data[target].unique())==1: 
        new_node=Node()
        new_node.leaf_node=True
        new_node.ans=data[target].unique()[0]
        return new_node

    best_gain=float('-inf')
    best_feat=None
    for feature in data.columns: 
        if feature==target:
            continue
        curr_gain=calc_info_gain(data,target, target)
        if best_gain<curr_gain: 
            best_gain=curr_gain
            best_feat=feature
            
    new_node=Node()
    new_node.attribute=best_feat
    
    for x in data[best_feat].unique(): 
        new_data=data[data[best_feat]==x].drop(best_feat,axis=1)
        new_node.branches[x]=recursive_split(new_data,target)

    return new_node 
```

```python
tree=ID3DecisionTree()
tree.root=recursive_split(data,'Diagnosis')

```

```python
def traverse_using_dfs(curr, u_age,u_bp,u_chol):
    if curr.leaf_node: 
        return curr.ans

    if curr.attribute=='Age' :
        return traverse_using_dfs(curr.branches[u_age],u_age, u_bp, u_chol)
    if curr.attribute=='Blood Pressure' :
        return traverse_using_dfs(curr.branches[u_bp],u_age, u_bp, u_chol)
    if curr.attribute=='Cholesterol' :
        return traverse_using_dfs(curr.branches[u_chol],u_age, u_bp, u_chol)
    
```
